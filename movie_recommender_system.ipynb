```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
from ast import literal_eval
mov_df=pd.read_csv("https://drive.google.com/uc?id=1rPR-P45M2UWsbXc8vpyCzWcQAYUfgVJX")
mov_df
```

```python
mov_df.info()
```

```python
new_df=mov_df[["genres","id","popularity","title","vote_average","vote_count"]]
new_df
```

```python
new_df[new_df.duplicated("id",keep="first")].sort_values(by="id")
```

```python
new_df=new_df.drop_duplicates(subset="id")
new_df[new_df.duplicated("id",keep="first")].sort_values(by="id")
```

```python
new_df.info()
```

```python
new_df.isna().sum()
```

```python
new_df.dropna(inplace=True)
new_df.info()
```

```python
new_df["id"]=new_df["id"].astype(int)
new_df["popularity"]=new_df["popularity"].astype(float)
new_df.info()
```

```python
new_df.head()
```

```python
new_df["genres"]=new_df["genres"].apply(lambda x: [i["name"] for i in literal_eval(x)])
new_df
```

```python
genre_df=new_df.explode("genres", ignore_index=False)
genre_df
```
